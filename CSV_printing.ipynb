```python
import serial
import csv
from datetime import datetime

ser = serial.Serial('COM4', 115200)

# Open CSV file for writing
# This creates a file called 'finger_data.csv'
with open('finger_data.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Timestamp", "Little", "Ring", "Middle", "Index", "Thumb"])

    try:
        while True:
            if ser.in_waiting:
                # Read the line sent from Arduino
                line = ser.readline().decode('utf-8').strip()
                
                # Get current computer time for the timestamp
                timestamp = datetime.now().isoformat()
                
                # Split the Arduino data (which is comma-separated) into a list
                data_points = line.split(',')
                
                # Combine the computer timestamp with the data from the Arduino
                row = [timestamp] + data_points
                
                # Save to the CSV file
                writer.writerow(row)
                
                # Print to your screen so you can see it's working
                print(f"{timestamp}, {line}")

    except KeyboardInterrupt:
        print("Stopped.")
        ser.close()
```
